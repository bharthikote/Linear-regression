```python
advertising  = pd.read_csv('/kaggle/input/advertising-dataset/advertising.csv')
```

```python
import seaborn as sns
```

```python
sns.pairplot(advertising, x_vars=['TV', 'Radio', 'Newspaper'], y_vars='Sales', size=4, kind='scatter')
plt.show()
```

```python
X = pd.DataFrame(advertising[ 'TV' ])
y = pd.DataFrame(advertising[ 'Sales' ])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( X, y, train_size = 0.7, test_size = 0.3, random_state = 100 )
```

```python
print( X_train.shape )
print( X_test.shape )
print( y_train.shape )
print( y_test.shape )
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit( X_train, y_train)
```

```python
#get intercept
print( model.intercept_ )
#get slope
print( model.coef_ )
```
